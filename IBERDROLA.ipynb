```python
from datetime import date
from dateutil import relativedelta

import pandas as pd

from pandas_datareader import data as pdr
import yfinance

import matplotlib.pyplot as plt
import seaborn as sns

yfinance.pdr_override()
```

```python
ticker = 'IBE.MC'
```

```python
end = date.today()
start = end - relativedelta.relativedelta(years=20)

values = pdr.get_data_yahoo(ticker, start, end)[['Close']]
```

```python
values.tail()
```

```python
values.plot(figsize=(20, 15))
plt.show()
```

```python
variation = values.pct_change()*100.0
variation.tail(50).plot(subplots=True, figsize=(20, 15))
plt.show()
```
